**Q2 - What is an interpreter in Python?**
- A python interpreter is a program that directly executes the python code, which means code can be executed as soon as it is written. In simple words, interpreter in Python processes the code line by line translating into a lower-level representation and simplify the development process.


**Q3 - What are pre-defined keywords in Python?**
- Pre-defined keywords are nothing but the reserved words in Python that holds a specific meaning behind them. These keywords are fixed set of words recognized by the Python interpreter. Each keyword serves a specific purpose like defining control flow (if, else, while or for), defining function and classes (def class), representing boolean values(True or False). These pre-defined keywords can not be redefined and these are case sensitive.


**Q4 - Can keywords be used as variable names?**

- No, keywords are pre-defined words with special meaning in programing language and can not be redefined, uning keyword as varialbe name will cause a syntax error.


**Q5 - What is mutability in Python?**
- Mutability refers to objects whose internal state or value can be changed after they have been created. It means, you can modify the content or the structure of a mutable object without creating a new object in memory. List[] is the perfect example of mutability that is ordered and changeable collection of items.

**Q6- Why are lists mutable, but tuples are immutable?**
- As far as I know, lists are mutable by design and the intent was clear to allow programmer to make any required changes after it's been created and tuples are immutable by design to prevent accidentally mutate the data that was intended to be immutable.

**Q7 - What is the difference between “==” and “is” operators in Python?**
- In python, the “==” and “is” operators are both used for comparision, but they evaluate the different aspects of object. "==" equality operator checks for value equality and "is" identity operator checks for object identity.

**Q8 - What are logical operators in Python?**
- There are many operators we use in Python but logical operators (and , or, not) are used to combine conditional statements, allowing you to perform operations based on multiple conditions.

**Q9 -  What is type casting in Python?**
- Type casting or type conversion in Python is a concept of converting one data type to another, there are two types of type casting, implicit and explicit type casting.


 **Q10 - What is the difference between implicit and explicit type casting?**
 - **Implicit type casting** is done automatically by Python and it is safe.
- **Explicit type casting** allows user to manually change the data type by using in-built functions, but it may cause data loss (Ex- converting float value (15.5) into int (15).

**Q11 -  What is the purpose of conditional statements in Python?**
- In Python, conditional statements are used to control the flow of execution in a program by making decisions based on certain conditions. In simple language it is basically to execute one condition when we have multiple condition in our program.

**Q12- How does the elif statement work?**
- The elif statement in Python allows you to check multiple conditions in a sequence, executing the block of code associated with the first true condition. The elif keyword stands for "else if" and is used in conditional statements to provide additional conditions to check if the initial if condition is false.

 **Q13- What is the difference between for and while loops?**
- Both for and while loops are control flow structures in programming that allow repeated execution of a block of code. However, they differ in their syntax, use cases, and behavior.

  - The for loop is used when the number of iterations is known beforehand or when iterating over a sequence (e.g., a list, range, or string). It combines initialization, condition checking, and iteration in a single structure.

  - The while loop is used when the number of iterations is not known in advance and depends on a condition. It continues as long as the condition evaluates to True.

 **Q14- Describe a scenario where a while loop is more suitable than a for loop.**
 - As we know that while loop is used when the number of iterations is not predetermined and depends on a condition being met during runtime. Unlike a for loop, which is ideal for iterating over a fixed range or sequence.
 While loop is more suitable in a situation where we want to give n numbers of warning to the user util user input the correct data.

```python
 #Q7- Write a program that performs basic arithmetic operations on two user-input numbers.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

# Handle division carefully to avoid division by zero
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (cannot divide by zero)"

print("\nResults:")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
```

```python
#Q14- Write a program to reverse a string using a while loop.
my_str="Mohd Bilal, I'am a Data science student"
length = len(my_str)
index = length - 1
reverse_str=""
while index >=0:
  reverse_str = reverse_str + my_str[index]
  index = index - 1
print(reverse_str)
```

```python
#Q15- Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num_f=int(input("Please enter a number you want the factorial of: "))
fact = 1
con = num_f
while con >=1:
  fact=fact * con
  con = con -1
print("factorial of", num_f, "is", fact)
```
